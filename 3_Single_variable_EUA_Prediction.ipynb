```python
#Import library
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# 데이터 로드
data = pd.read_csv("Monthly_EUA2.csv")
```

```python
# 날짜를 인덱스로 설정
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.set_index('Timestamp', inplace=True)
```

```python
# ARIMA 모델 피팅 #Gird search를 통한 최적의 p, q, d 선정하였음
model = ARIMA(data, order=(30,1,2))
fit_model = model.fit()
```

```python
# 미래 예측
future_steps = 36  # 예측할 미래 단계 수
forecast = fit_model.forecast(steps=future_steps)
```

```python
# 결과 시각화
plt.figure(figsize=(7, 4))
plt.plot(data, 'black', label='history',linewidth=3)
plt.plot(forecast, 'b', label='future prediction',linewidth=3)
plt.legend(fontsize = "15")
plt.grid(True)
plt.show()
```
